```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr

# Install necessary dependencies (Uncomment in Colab)
# !pip install torch torchvision torchaudio diffusers gradio

def load_model():
    """Load the Stable Diffusion model."""
    model_id = "stabilityai/stable-diffusion-2"
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    pipe.to("cuda" if torch.cuda.is_available() else "cpu")
    return pipe

# Load the model
pipe = load_model()

def generate_image(prompt, inference_steps, guidance_scale):
    with torch.inference_mode():
        image = pipe(prompt, num_inference_steps=inference_steps, guidance_scale=guidance_scale).images[0]
    return image

# Create a Gradio Interface
interface = gr.Interface(
    fn=generate_image,
    inputs=[
        gr.Textbox(label="Enter a clothing description", placeholder="e.g., 'red dress', 'modern suit', 'summer outfit'"),
        gr.Slider(minimum=10, maximum=100, value=50, label="Inference Steps"),
        gr.Slider(minimum=1.0, maximum=15.0, value=7.5, label="Guidance Scale")
    ],
    outputs=gr.Image(type="pil"),
    title="CoutureAI: Clothing Image Generator Using Stable Diffusion Pipeline",
    description="Generate AI-powered clothing designs using Stable Diffusion."
)

# Launch the app
interface.launch(debug=True, share=True)
```
